```python
def security(y):
    if(y['Securities Account'] == 1) & (y['CD Account'] == 1):
        return "Holds Securities & Depostie"
    if(y['Securities Account'] == 0) & (y['CD Account'] == 0):
        return "Does not hold Securites & Depsotie account"
    if(y['Securities Account'] == 1) & (y['CD Account'] == 0):
        return "Holds only Scurites account"
    if(y['Securities Account'] == 0) & (y['CD Account'] == 1):
        return "Holds only Deposite account"
```

```python
data['Account_holder_category'] = data.apply(security, axis = 1)
```

```python
def plot(col1, col2, lable1, lable2, title):
    sns.distplot(data[data[col2] == 0][col1], hist = False, 
            label =lable1)
    
    sns.distplot(data[data[col2] == 1][col1], hist = False, 
            label = lable2)
    
    plt.legend()
    plt.title(title)
```

```python
plot('Income', 'Personal Loan', "Income with no persoanl loan", "Income with persoanl loan", "Income Disrbution plot")
```

- **Conclusion**    Through this project, I explored and analyzed customer data from a personal loan campaign by Thera Bank. I learned how different features like income, education, and credit card usage can influence whether a customer accepts a loan offer.

Some key insights I found:

Customers with higher income are more likely to accept personal loans.

Credit card spending is another important factor — people who spend more are more interested in loans.

Education also plays a role. Professionals and graduates are more likely to take loans compared to undergraduates.

Customers who have both securities and CD accounts are more open to personal loan offers.
